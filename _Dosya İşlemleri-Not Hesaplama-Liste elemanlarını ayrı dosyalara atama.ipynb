```python
## Programlama Ödevi -  Dosya İşlemleri

### Proje 1 

#Bir sınıfın harf notlarını hesapladığımız programı geliştirerek 
#kalanları "kalanlar.txt" dosyasında ve geçenleri "geçenler.txt" dosyasına yazmaya çalışın.
```

```python
def not_hesapla(satır):
     satır = satır[:-1]
     
     liste = satır.split(",")
     
        
     isim = liste[0]
     not1 = int(liste[1])
     not2 = int(liste[2])
     not3 = int(liste[3])
     
     son_not = not1 * (1/3) + not2 * (1/3) + not3 * (1/4)
     
     if (son_not >= 90):
        harf = "AA"
     elif (son_not >= 85):
        harf = "BA"
     elif (son_not >= 80):
        harf = "BB"
     elif (son_not >= 75):
        harf = "CB"
     elif (son_not >= 70):
        harf = "CC"
     elif (son_not >= 65):
        harf = "DC"
     elif (son_not >= 60):
        harf = "DD"
     elif (son_not >= 55):
        harf = "FD"
     else:
        harf = "FF"    
        
     return harf
```

```python
with open("C:/Users/burcu/OneDrive/Masaüstü/Ek_Calısma/Python MMC/Yeni klasör/Sifirdan-Ileri-Seviyeye-Python-Programlama-master/Dosya İşlemleri/Kodlama Egzersizleri/dosya.txt","r",encoding= "utf-8") as file:
   kalanlar =[]
   gecenler =[]
   for i in file :
      if  not_hesapla(i) == "FF" :
         kalanlar.append(i)
      else:
        gecenler.append(i)
    
```

```python
with open("gecenler.txt","r+",encoding="utf-8") as file2:
    for i in gecenler:
       file2.write(i)
       print(file2.read())
```

```python
with open("kalanlar.txt","r+",encoding="utf-8") as file3:
    for i in gecenler:
       file3.write(i)
       print(file3.read())
```
